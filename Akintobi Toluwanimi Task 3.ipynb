## Pizza Place Sales

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Importing the datasets

pizzas = pd.read_csv(r"C:\Users\Toluwanimi\Downloads\Pizza+Place+Sales\pizza_sales\pizzas.csv")
orders = pd.read_csv(r"C:\Users\Toluwanimi\Downloads\Pizza+Place+Sales\pizza_sales\orders.csv")
pizza_types = pd.read_csv(r"C:\Users\Toluwanimi\Downloads\Pizza+Place+Sales\pizza_sales\pizza_types.csv")
order_details = pd.read_csv(r"C:\Users\Toluwanimi\Downloads\Pizza+Place+Sales\pizza_sales\order_details.csv")
```

```python
#Importing the data dictionary

data_dictionary = pd.read_csv(r"C:\Users\Toluwanimi\Downloads\Pizza+Place+Sales\data_dictionary.csv")

data_dictionary
```

```python
#The first 5 rows of the pizzas dataset

pizzas.head()
```

```python
#The first 5 rows of the pizza_types dataset

pizza_types.head()
```

```python
#The first 5 rows of the orders dataset

orders.head()
```

```python
#The first 5 rows of the order_details dataset

order_details.head()
```

```python
#Merging the pizzas and pizza_types datasets together

data_1 = pd.merge(pizzas, pizza_types, on="pizza_type_id", how="inner")

data_1.head()
```

```python
#Merging the orders and order_details datasets together

data_2 = pd.merge(orders, order_details, on="order_id", how="inner")

data_2.head()
```

```python
#Joining the two merged datasets together

data = pd.merge(data_1, data_2, on="pizza_id", how="inner")

data
```

```python
#The first 5 rows of the fully merged dataset

data.head()
```

```python
#The number of rows and columns of the dataset

data.shape
```

```python
#Full info about the dataset

data.info()
```

```python
#Checking for null values

data.isna().sum()
```

```python
#Converting the date and time columns to the right datatype

data["date"] = pd.to_datetime(data["date"])

data["time"] = pd.to_datetime(data["time"])
```

```python
data.info()
```

### Now that there are no null values and all the columns are in the right datatype, analysis can begin.

```python
data.head()
```

```python
#Total revenue/sales

revenue = (data["price"] * data["quantity"]).sum()

print(f"The Total Revenue/Sales is {revenue}")
```

```python
#Total Quantity Sold

quantity = data["quantity"].sum()

print(f"The Total Quantity sold is {quantity}")
```

```python
#Total Orders

orders = data["order_id"].nunique()

print(f"The Total number of orders is {orders}")
```

```python
#The pizza types

pizza_types = data["pizza_type_id"].unique()
pizza_types
```

```python
#Number of pizza types they sell

pizza_types_no = data["pizza_type_id"].nunique()
print(f"They sell {pizza_types_no} pizza types")
```

```python
#Average price of the pizza

avg_price = data["price"].mean()
print(f"The average price of the pizzas is {avg_price}")
```

```python
#The peak hours of sales

data["hour"] = data["time"].dt.hour  #Adding a new hour column
```

```python
data["revenue"] = data["price"] * data["quantity"]  #Adding a new revenue column
```

```python
data.head()
```

```python
#Grouping the hours in terms of revenue

sales_hours = data.groupby("hour")["revenue"].sum().reset_index()

sales_hours
```

```python
sales_hours = sales_hours.sort_values(by = "revenue", ascending=False)

print("The total sales made in each hour are:")
sales_hours
```

```python
#A bar chart showing the hourly sales

plt.bar(sales_hours["hour"], sales_hours["revenue"], color="purple")
plt.title("Total Sales by each hour of the day")
plt.xlabel("Hour")
plt.ylabel("Revenue")

plt.show()
```

### Based on this analysis, the peak hours of sales are between 12 PM and 1 PM, which corresponds to lunchtime. This means that the pizza place gets the most revenue during midday.

```python
#Total Sales made on each day of the week

data["day"] = data["date"].dt.day_name()  #Adding a new day column
```

```python
data.head()
```

```python
#Grouping the days in terms of revenue

sales_days= data.groupby("day")["revenue"].sum().reset_index()

print("The total sales made on each day of the week are:")
sales_days
```

```python
#The day sales are made the most

sales_days = sales_days.sort_values(by = "revenue", ascending=False)

sales_days
```

```python
#A bar chart showing the total daily sales

plt.bar(sales_days["day"], sales_days["revenue"], color="purple")
plt.title("Total Sales by each day of the week")
plt.xlabel("Day")
plt.xticks(rotation=45)
plt.ylabel("Revenue")

plt.show()
```

### This shows that Friday is when sales are made the most, suggesting that customers tend to order more pizzas towards the end of the work week.

```python
#The top 5 bestselling pizzas

top_pizzas = data.groupby("pizza_id")["quantity"].sum().reset_index().sort_values(by="quantity", ascending=False).head()

print("The top 5 bestselling pizzas are:")
top_pizzas
```

```python
#The sales made in each month

data["month"] = data["date"].dt.month_name()  #Adding a new month column
data.head()
```

```python
#Grouping the months in terms of revenue

monthly_sales = data.groupby("month")["revenue"].sum().reset_index().sort_values(by="revenue", ascending=False)

print("The sales made in each month are:")
monthly_sales
```

```python
#A bar chart showing the total monthly sales

plt.bar(monthly_sales["month"], monthly_sales["revenue"], color="purple")
plt.title("Total Sales made in each month")
plt.xlabel("Month")
plt.xticks(rotation=45)
plt.ylabel("Revenue")

plt.show()
```

### The pizza place has it's highest sales in July and May, and the lowest sales in September and October. Overall, there is not a consistent upward or downward trend across the months. Instead, the sales fluctuate.

```python
data.head()
```

```python
#Total sales made by each pizza type on the menu

pizza_sales = data.groupby("name")["revenue"].sum().reset_index().sort_values(by="revenue", ascending=True)

pizza_sales
```

```python
#A bar chart showing each pizza sales

plt.bar(pizza_sales["name"], pizza_sales["revenue"], color="purple")
plt.title("Total sales made by each pizza type on the menu")
plt.xlabel("Pizza Name")
plt.xticks(rotation=90)
plt.ylabel("Revenue")

plt.show()
```

### This shows that these pizzas, such as The Brie Carre Pizza and The Green Garden Pizza, are not doing well on the menu in terms of revenue.
